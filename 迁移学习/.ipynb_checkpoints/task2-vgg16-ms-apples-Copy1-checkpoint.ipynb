```python
#load the data
from keras.preprocessing.image import load_img,img_to_array

img_path = '1.jpg'
img = load_img(img_path,target_size=(224,224))
img = img_to_array(img)

type(img)
```

```python
#visualize the data
%matplotlib inline
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(5,5))
img = load_img(img_path,target_size=(224,224))
plt.imshow(img)
```

```python
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
import numpy as np
model_vgg = VGG16(weights='imagenet',include_top=False)
x = np.expand_dims(img,axis=0)
x = preprocess_input(x)
print(x.shape)
```

```python
#特征提取
features = model_vgg.predict(x)
print(features.shape)
```

```python
#flatten
features = features.reshape(1,7*7*512)
print(features.shape)
```

```python
#list file names of the training datasets
import os
folder = "train_data"
dirs = os.listdir(folder)
#generate path for the images
img_path = []
for i in dirs:                             
    if os.path.splitext(i)[1] == ".jpg":   
        img_path.append(i)
img_path = [folder+"//"+i for i in img_path]
#print(img_path)
```

```python
#define a method to load and preprocess the image
def modelProcess(img_path,model):
    img = load_img(img_path, target_size=(224, 224))
    img = img_to_array(img)
    x = np.expand_dims(img,axis=0)
    x = preprocess_input(x)
    x_vgg = model.predict(x)
    x_vgg = x_vgg.reshape(1,7*7*512)
    return x_vgg
```

```python
#preprocess multiple images
features1 = np.zeros([len(img_path),7*7*512])
for i in range(len(img_path)):
    feature_i = modelProcess(img_path[i],model_vgg)
    print('preprocessed:',img_path[i])
    features1[i] = feature_i
```

```python
X = features1
```

```python
from sklearn.cluster import KMeans

cnn_kmeans = KMeans(n_clusters=2,max_iter=2000)
cnn_kmeans.fit(X)
y_pred_kmeans = cnn_kmeans.fit_predict(X)
```

```python
print(y_pred_kmeans)
```

```python
from collections import Counter
print(Counter(y_pred_kmeans)) # 调用Counter函数
```

```python
folder_test = "test_data"
dirs_test = os.listdir(folder_test)
print(dirs_test)
#generate path for the images
img_path_test = []
for i in dirs_test:                             
    if os.path.splitext(i)[1] == ".jpg":   
        img_path_test.append(i)
img_path_test = [folder_test+"//"+i for i in img_path_test]
print(img_path_test)
#preprocess multiple images
features_new_test = np.zeros([len(img_path_test),7*7*512])
for i in range(len(img_path_test)):
    feature_i = modelProcess(img_path_test[i],model_vgg)
    print('preprocessed:',img_path_test[i])
    features_new_test[i] = feature_i

y_predict = cnn_kmeans.predict(features_new_test)
print(y_predict)
```

```python
from sklearn.preprocessing import StandardScaler
stds = StandardScaler()
X_norm = stds.fit_transform(X)
#pca analysis
from sklearn.decomposition import PCA
pca = PCA(n_components=200)
X_pca = pca.fit_transform(X_norm)
#calculate the variance ratio of each principle components
var_ratio = pca.explained_variance_ratio_
print(np.sum(var_ratio))
print(X_pca.shape)
#from sklearn.utils import shuffle
#X1 = shuffle(X_pca)

X1 = X_pca
```

```python
#try meanshift model
from sklearn.cluster import MeanShift,estimate_bandwidth
#obtain the bandwidth
bw = estimate_bandwidth(X1,n_samples=160)
print(bw)
#establish the meanshift model-un-supervised model
ms_pca = MeanShift(bandwidth=bw)
ms_pca.fit(X1)
y_pred_ms_pca = ms_pca.predict(X_pca)
```
